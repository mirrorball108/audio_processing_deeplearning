```python
import librosa, librosa.display
import matplotlib.pyplot as plt
```

```python
file='/content/audio.mp3'
signal,sr=librosa.load(file,sr=22050)
#signal= sr*duration of the song
```

```python
#waveform
librosa.display.waveshow(signal,sr=sr)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

```python
#fft ->spectrum
import numpy as np

fft=np.fft.fft(signal)
magnitude=np.abs(fft)
#magnitude plots the contribution of every individual frequency

frequency=np.linspace(0,sr, len(magnitude))
#linspace returns evenly spaced numbers in an interval (0,sr)

plt.plot(frequency,magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()
```

the graph is symmetrical


this is a property of fourier transform


only the left half gives us novel information

```python
left_frequency=frequency[:int(len(frequency)/2)]
left_magnitude=magnitude[:int(len(magnitude)/2)]

plt.plot(left_frequency,left_magnitude)
plt.xlabel("'eft_Frequency")
plt.ylabel("left_Magnitude")
plt.show()
```

```python
#stft ->spectrogram
n_fft=2048
hop_length=512

stft=librosa.core.stft(signal,hop_length=hop_length,n_fft=n_fft)
spectrogram=np.abs(stft)
librosa.display.specshow(spectrogram,sr=sr,hop_length=hop_length)
plt.xlabel("time")
plt.ylabel("frequency")
#the amplitude is expressed through colour
plt.colorbar()
plt.show()
```

```python
log_spectrogram=librosa.amplitude_to_db(spectrogram)
librosa.display.specshow(log_spectrogram,sr=sr,hop_length=hop_length)
plt.xlabel("time")
plt.ylabel("frequency")
#the amplitude is expressed through colour
plt.colorbar()
plt.show()
```

**MFCC**
MEL FREQUENCY CEPSTRAL COEEFICIENTS(MFCCs)

1.capture timbral/textural aspects of sound

2.frequency domain feature(not time)

3.approximate human auditory system(try to mimic the way humans perceive sound)

4. used in- speech recognition, music genre classification, music instrument classification


```python
# MFCCs
MFCC=librosa.feature.mfcc(y=signal,n_fft=n_fft,hop_length=hop_length,n_mfcc=13)
librosa.display.specshow(MFCC,sr=sr,hop_length=hop_length)
plt.xlabel("time")
plt.ylabel("MFCC")
#the amplitude is expressed through colour
plt.colorbar()
plt.show()

```
